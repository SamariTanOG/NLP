```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import  nltk
import re
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from  sklearn.feature_extraction.text import CountVectorizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import pandas as pd
inp = 'C:/Users/arpit/Downloads/Input.xlsx'
data = pd.read_excel(inp)
```

```python
def extract_article_text(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for HTTP errors (404, 500, etc.)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Customize this part based on the structure of the article content
        article_text_tags = soup.find_all('p')
        article_text = '\n'.join([tag.get_text() for tag in article_text_tags])
        
        return article_text
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {e}")
        return None
```

```python
def save_article_to_file(article_text, url_id, output_directory):
    if article_text is not None:
        file_name = f"{url_id}.txt"
        file_path = os.path.join(output_directory, file_name)
        with open(file_path, 'w', encoding='utf-8') as f:
            f.write(article_text)
        print(f"Article text saved to {file_path}")
    else:
        print(f"No article text to save for url_id {url_id}")
```

```python
output_dir = "C:/Users/arpit/OneDrive/Desktop/Data/"
os.makedirs(output_dir, exist_ok=True)
```

```python
for index, row in data.iterrows():
    url_id = row['URL_ID']
    article_html = row['URL']
    
    article_text = extract_article_text(article_html)
    
    
    save_article_to_file(article_text, url_id, output_dir)
print("Extraction completed.")
```

```python
file_list = os.listdir(output_dir)
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
sscore = []
```

```python
for name in file_list:
    
    if os.path.isfile(os.path.join(output_dir, name)):
        file_path = os.path.join(output_dir,name)
        with open(file_path, 'r', encoding='utf-8') as f:
            f.read()
            sentiment_scores = analyzer.polarity_scores(file_path)
        sscore.append(sentiment_scores)

        
```
